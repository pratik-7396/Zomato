```python
df3.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
df3.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
rating = df3.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
rating
```

```python
Ratings = rating.rename(columns={0:'Count'})
```

```python
Ratings
```

Observations:
    1.From total observations 2148 has no rated the service of Zomato.
    2.When rating is in between 1.8 to 2.4 indicates that the rating is poor
    3.When rating is in between 2.5 to 3.4 indicates that the rating is Average
    4.When rating is in between 3.5 to 3.9 indicates that the rating is Good.
    5.When rating is in between 4.0 to 4.4 indicates that the rating is Very Good.
    6.When rating is in between 4.5 to 4.9 indicates that the rating is Excellent.

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x= 'Aggregate rating', y= 'Count',data=ratings)
```
